# Why Python Belongs on the Lab Bench
### 1.1 — Foundations · Beginner · 10–12 min

You already trust your instrument. The question this track answers is how to trust
what happens *after* the instrument — when the numbers leave the vendor software and
you have to turn them into a result.

Vendor software is built for one screen and one result. The moment you want to do the same calculation across twenty samples, double-check the numbers, or hand the work to a colleague who can reproduce it, you’ve outgrown the menus. Python is simply a more transparent and reproducible way to work — and you don’t need to “learn programming” to get value from it today.

> **One idea to hold onto:** a notebook is a lab notebook that also does the
> arithmetic — every number you see was produced by a step you can read, check, and
> re-run.

**By the end of this notebook you will have:**

1. Run code in a notebook and seen a result appear.
2. Turned three replicate readings into a reported mean and %RSD.
3. Converted a measured absorbance into a concentration with Beer–Lambert.

## 1. Your first result: replicate readings

Here are five replicate absorbance readings of one standard. We'll store them and
ask Python for the mean and the precision (%RSD) — the same summary you'd report.

Run the cell (Shift+Enter). Everything after the `#` is a comment for you, ignored
by Python.

```python
import numpy as np  # numpy: fast math on lists of numbers

# Five replicate absorbance readings (AU) of the same standard.
absorbance = [0.451, 0.448, 0.455, 0.450, 0.447]

mean_abs = np.mean(absorbance)            # average reading
sd_abs = np.std(absorbance, ddof=1)       # sample standard deviation (ddof=1)
rsd_pct = 100 * sd_abs / mean_abs         # relative standard deviation, in %

print("mean absorbance :", round(mean_abs, 4), "AU")
print("std deviation   :", round(sd_abs, 4), "AU")
print("%RSD            :", round(rsd_pct, 2), "%")
```

That's the whole idea. You ran a step, and a result appeared — and you can see
*exactly* how it was computed. Nothing is hidden behind a button. A %RSD under ~1%
on replicates like these is the kind of precision check you'd do before trusting a
reading.

## 2. From signal to chemistry: Beer–Lambert in one line

A reading in AU isn't the answer a chemist wants — concentration is. Beer–Lambert
gives it: **A = ε · b · c**, so **c = A / (ε · b)**, where ε is the molar
absorptivity and b is the path length.

```python
epsilon = 6200.0          # molar absorptivity, L / (mol * cm)  -- from the method
path_length_cm = 1.0      # standard 1 cm cuvette

# Use the mean absorbance from section 1.
concentration_mol_per_L = mean_abs / (epsilon * path_length_cm)

# Report in micromolar (1 mol/L = 1e6 umol/L) -- a friendlier unit here.
concentration_umol_per_L = concentration_mol_per_L * 1e6

print("concentration:", round(concentration_umol_per_L, 2), "umol/L")
```

You just went from raw signal to a reported concentration in two readable steps.
Imagine doing that for a tray of 96 samples — that's where this is heading, and
it's why the vendor menus stop being enough.

## 3. See it

A quick plot of the five replicates with the mean drawn through them. Seeing the
scatter around the mean is your visual precision check.

```python
import matplotlib.pyplot as plt

replicate_number = [1, 2, 3, 4, 5]

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(replicate_number, absorbance, label="readings")
ax.axhline(mean_abs, linestyle="--", label="mean")
ax.set_xlabel("replicate #")
ax.set_ylabel("absorbance (AU)")
ax.set_title("Five replicate readings of one standard")
ax.legend(fontsize=8)
plt.tight_layout()
plt.show()
```

## Key Takeaways

- A notebook runs code in **cells**; each result is produced by a step you can read.
- `numpy` gives you `mean`, `std`, and friends for everyday lab summaries.
- A measured **absorbance** becomes a **concentration** through Beer–Lambert — and
  every step is visible, not hidden in a menu.

## Practical Checklist

- [ ] Run a cell with **Shift+Enter** and read the printed result.
- [ ] Compute a mean and %RSD for a set of replicates.
- [ ] Keep units in your reported numbers (AU, umol/L), not just bare digits.

## Common Mistakes

- Using population SD where you want **sample** SD — pass `ddof=1` for replicates.
- Reporting a bare number with no unit; a value without a unit isn't a result.
- Expecting code to run "later in the file" — a cell only runs when you run it
  (the subject of the next lesson).
