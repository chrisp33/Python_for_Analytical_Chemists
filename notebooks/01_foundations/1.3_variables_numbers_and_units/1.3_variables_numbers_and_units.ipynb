# Variables, Numbers, and Units Without Tears
### 1.3 — Foundations · Beginner · 10–12 min

Most wrong results in a lab are not wrong measurements — they're unit slips. A
milligram read as a gram, a millilitre treated as a litre, and the number is off by
1000 with nothing to warn you.

Python can't know your units. But you can write your variables so that a unit mistake
*looks* wrong on the page — and that visibility is most of the protection.

> **One idea to hold onto:** put the unit in the variable name. `volume_mL` and
> `volume_L` can never be silently confused if you always say which one you mean.

**By the end of this notebook you will be able to:**

1. Store measured quantities in clearly named, unit-tagged variables.
2. Do dilution and molarity arithmetic with explicit conversions.
3. Recognize the classic silent unit error — and the habit that prevents it.

## 1. Name the quantity, name the unit

We'll prepare a standard: weigh out a solid, dissolve to a known volume, and compute
the molarity. Watch how every variable name carries its unit.

```python
# Inputs, each tagged with its unit in the name.
mass_mg = 25.0                 # mass weighed out, milligrams
molar_mass_g_per_mol = 158.03  # e.g. KMnO4, grams per mole
volume_mL = 250.0              # final flask volume, millilitres

# Convert to base SI-friendly units BEFORE the chemistry, explicitly.
mass_g = mass_mg / 1000.0      # mg -> g
volume_L = volume_mL / 1000.0  # mL -> L

moles = mass_g / molar_mass_g_per_mol         # mol
molarity_mol_per_L = moles / volume_L         # mol / L

print("moles     :", round(moles, 6), "mol")
print("molarity  :", round(molarity_mol_per_L * 1000, 4), "mmol/L")
```

Every conversion is a visible line of its own (`mg -> g`, `mL -> L`). That's
deliberate: a reader — including future you — can check each one. The chemistry only
happens *after* everything is in consistent units.

## 2. The silent unit mistake

Here's the error that bites real labs. Suppose we forget to convert the volume from
millilitres to litres and divide by `volume_mL` directly.

```python
# WRONG: moles divided by volume in mL gives mol/mL, not mol/L.
molarity_wrong = moles / volume_mL          # forgot the mL -> L step

# RIGHT: divide by volume in litres.
molarity_right = moles / volume_L

print("wrong (mol/mL mistaken for mol/L):", round(molarity_wrong * 1000, 6), "mmol/L")
print("right (true mol/L)               :", round(molarity_right * 1000, 4), "mmol/L")
print("off by a factor of               :", round(molarity_right / molarity_wrong))
```

A factor of **1000**, and Python raised no error — the arithmetic was valid, the
*units* were not. No tool catches this for you. The defenses are entirely habit:
convert in named steps, keep units in your variable names, and sanity-check the
magnitude (a benchtop standard is rarely micromolar *or* molar by accident).

## 3. Whole numbers, decimals, and reporting

Two small things that trip up beginners:

- A count (like a replicate number) is an **integer**; a measurement is a **float**
  (it has a decimal point). Write measured values with a decimal — `25.0`, not `25` —
  so they behave like measurements.
- When you report, format the number to a sensible precision and **state the unit**.

```python
dilution_factor = 20          # an integer: a 20-fold dilution
measured_umol_L = 3.7         # a float: a measurement

stock_umol_L = measured_umol_L * dilution_factor

# f-strings let you format value and unit together, to a chosen precision.
print(f"diluted reading : {measured_umol_L:.2f} umol/L")
print(f"x{dilution_factor} dilution    : {stock_umol_L:.1f} umol/L in the stock")
```

## Key Takeaways

- **Put the unit in the variable name** — `mass_mg`, `volume_L` — so mismatches show.
- Convert units in **explicit, named steps** before doing the chemistry.
- The classic silent error is a missing `mL -> L` (or `mg -> g`) conversion: valid
  arithmetic, wrong units, no error message.

## Practical Checklist

- [ ] Every measured variable name ends in its unit.
- [ ] Unit conversions are their own clearly labelled lines.
- [ ] Reported numbers are formatted to sensible precision **and** carry a unit.
- [ ] You sanity-checked the magnitude against what you expect for that sample.

## Common Mistakes

- Mixing mL and L (or mg and g) in one calculation — a quiet factor-of-1000 error.
- Writing measurements as bare integers (`25`) instead of floats (`25.0`).
- Reporting a value with no unit, leaving the reader to guess.
