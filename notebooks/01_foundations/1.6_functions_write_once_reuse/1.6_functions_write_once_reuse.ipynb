# Functions: Writing a Step Once and Reusing It
### 1.6 — Foundations · Beginner · 12–15 min

In the lab, a method is defined once and then applied the same way to every sample,
every day. A **function** is how you do that in Python: you write the step a single
time, give it a name, and call it wherever you need it. Define it once, fix it in one
place, and trust that it behaves identically everywhere.

> **One idea to hold onto:** a function is a method definition — write the step once,
> and every call applies it the same way.

**By the end of this notebook you will be able to:**

1. Turn a repeated calculation into a named function with a clear input and output.
2. Give it a sensible default (like a 1 cm path length) and a one-line description.
3. Reuse it across a batch — combining the function with the loop from 1.5.

## 1. A step you keep repeating

Here's the Beer–Lambert conversion from earlier lessons, written out for three
samples. Notice the same formula copied three times — three chances to mistype it.

```python
epsilon = 6200.0
path_length_cm = 1.0

c1 = (0.451 / (epsilon * path_length_cm)) * 1e6
c2 = (0.902 / (epsilon * path_length_cm)) * 1e6
c3 = (0.121 / (epsilon * path_length_cm)) * 1e6

print(round(c1, 2), round(c2, 2), round(c3, 2), "umol/L")
```

If the method changes — a different path length, a unit tweak — you'd have to find and
fix every copy. That's exactly how inconsistencies sneak into an analysis.

## 2. Wrap it in a function

We define the step once with `def`. It takes inputs (the absorbance, and ε and path
length), and **returns** the concentration. The `path_length_cm=1.0` gives a sensible
default you can override when needed.

```python
def absorbance_to_concentration(absorbance, epsilon, path_length_cm=1.0):
    "Beer-Lambert: convert absorbance (AU) to concentration in umol/L."
    concentration_mol_L = absorbance / (epsilon * path_length_cm)
    return concentration_mol_L * 1e6   # mol/L -> umol/L

# Call it -- the 1 cm path length is assumed by the default.
one_result = absorbance_to_concentration(0.451, epsilon=6200.0)
print(round(one_result, 2), "umol/L")
```

The formula now lives in exactly **one place**. Every call runs the identical step,
and a method change is a one-line edit inside the function.

## 3. Reuse it across a batch

Functions and loops are a natural pair: the loop visits each sample, the function does
the work. We'll also write a second small function to apply the spec check from 1.5,
so each sample is converted *and* judged by reusable steps.

```python
def flag_result(concentration_umol_L, low, high):
    "Classify a concentration against spec limits: 'LOW', 'HIGH', or 'pass'."
    if concentration_umol_L < low:
        return "LOW"
    elif concentration_umol_L > high:
        return "HIGH"
    return "pass"

sample_ids = ["S-01", "S-02", "S-03", "S-04", "S-05"]
absorbance = [0.451, 0.902, 0.121, 0.660, 0.405]

for sid, ab in zip(sample_ids, absorbance):
    conc = absorbance_to_concentration(ab, epsilon=6200.0)   # reused
    flag = flag_result(conc, low=40.0, high=120.0)           # reused
    print(f"{sid}: {conc:6.2f} umol/L  ->  {flag}")
```

## 4. Why this matters

Two named steps — `absorbance_to_concentration` and `flag_result` — now define your
mini-method, and the loop applies them to every sample. If a reviewer asks "how did
you compute concentration?", the answer is one short, readable function, not a formula
scattered across cells.

This is the bridge to the rest of the course. The reusable preprocessing steps you'll
meet in Track 2 and beyond — baseline correction, smoothing, scatter correction — are
all just functions like these, written once and applied consistently.

## Key Takeaways

- A **function** defines a step once; every call applies it identically.
- **Default arguments** (like `path_length_cm=1.0`) capture the common case while
  staying overridable.
- Functions + loops = a method applied consistently across a whole batch.

## Practical Checklist

- [ ] When you copy-paste a calculation, turn it into a function instead.
- [ ] Give the function a clear name and a one-line description of what it returns.
- [ ] Use defaults for the usual values; pass explicit ones when they differ.
- [ ] Reuse the same function across samples rather than rewriting the step.

## Common Mistakes

- Copy-pasting a formula and editing only some of the copies.
- Functions that both compute *and* print — return the value, and let the caller
  decide what to do with it.
- Burying method constants inside the function body where they're hard to find;
  pass them in, or default them clearly.
