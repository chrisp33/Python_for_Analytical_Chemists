# Notebooks, Scripts, and How to Not Lose Your Work
### 1.2 — Foundations · Beginner · 8–10 min

A result you can't reproduce isn't a result yet — it's a number you happened to get
once. The good news: a notebook makes reproducibility almost free, *if* you treat it
the way you treat a lab notebook.

> **One idea to hold onto:** the real test of an analysis is **Restart & Run All** —
> if it produces the same answer from a clean start, you can defend it.

**By the end of this notebook you will be able to:**

1. Explain why cells run in the order *you* run them, not top to bottom by magic.
2. Re-run an analysis and confirm you get the identical number.
3. Decide when to keep something in a notebook and when to move it to a script.

## 1. Cells run in the order you run them

A notebook keeps a running memory (the *kernel*). A variable exists once you run the
cell that creates it — and *only* then. Here we blank-correct a reading: subtract the
blank from the sample.

```python
measured_absorbance = 0.512   # sample, AU
blank_absorbance = 0.034      # reagent blank, AU

corrected_absorbance = measured_absorbance - blank_absorbance
print("blank-corrected absorbance:", round(corrected_absorbance, 4), "AU")
```

If you had run the third line *before* the first two, Python wouldn't know what
`measured_absorbance` was — it would error. That's the one notebook gotcha: running
cells out of order leaves stale values in memory.

**The fix is a habit:** when an analysis is done, choose **Kernel → Restart & Run
All**. It clears memory and runs every cell from the top, exactly as a colleague
would. If it still works, your notebook is honest.

## 2. The reproducibility test: same inputs, same answer

Reproducible means: feed in the same numbers, get back the same result — every time.
Let's prove it. We compute a dilution-corrected concentration twice and check they
match.

```python
def corrected_concentration(raw_conc_umol_L, dilution_factor):
    # One small, explicit step: undo a dilution.
    return raw_conc_umol_L * dilution_factor

first_run = corrected_concentration(12.4, dilution_factor=10)
second_run = corrected_concentration(12.4, dilution_factor=10)

print("first run :", first_run, "umol/L")
print("second run:", second_run, "umol/L")
print("identical?:", first_run == second_run)
```

`True`. Same inputs, same output — that is what "reproducible" means in practice.
When a calculation depends on randomness (later lessons do), we'll fix a *seed* so
even the random parts repeat exactly.

## 3. Put your constants at the top

The easiest way to keep a notebook honest is to gather the values your analysis
depends on — path length, blank, dilution, spec limits — in one block near the top.
Then every later cell reads from one place, and changing a method value is a
one-line edit, not a hunt.

```python
# --- Method constants (edit here, used everywhere below) ---
PATH_LENGTH_CM = 1.0
BLANK_ABSORBANCE = 0.034
DILUTION_FACTOR = 10
# -----------------------------------------------------------

sample_absorbance = 0.512
net_absorbance = sample_absorbance - BLANK_ABSORBANCE
print("net absorbance:", round(net_absorbance, 4), "AU",
      "| path length:", PATH_LENGTH_CM, "cm",
      "| dilution:", DILUTION_FACTOR)
```

## 4. Notebook or script?

Both run the same Python. They serve different moments:

| Use a **notebook** when... | Use a **script** when... |
|---|---|
| You're exploring, plotting, deciding what to do | The method is settled and you just want it to run |
| You want results and figures inline with notes | You'll run it unattended, on many files, repeatedly |
| You're teaching or documenting an analysis | You're handing a fixed procedure to a pipeline |

Track 1 lives entirely in notebooks. Turning a finished notebook into a reusable
script is a Track 9 skill — by then you'll have a method worth scripting.

## Key Takeaways

- A notebook has a **memory (kernel)**; a variable exists only after you run its cell.
- **Restart & Run All** is the reproducibility test — make it a habit before you trust
  or share a notebook.
- Gather **method constants at the top** so the analysis has a single source of truth.

## Practical Checklist

- [ ] Before trusting results, run **Kernel → Restart & Run All** and confirm it works.
- [ ] Keep path length, blank, dilution, and spec limits in one constants block.
- [ ] Re-run a finished analysis and confirm the key number is unchanged.

## Common Mistakes

- Running cells out of order and trusting a **stale** value left in memory.
- Hard-coding the same constant in five cells, then changing only four of them.
- Treating "it ran once" as reproducible without a clean restart.
